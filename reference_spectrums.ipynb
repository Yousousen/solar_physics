Using the specified nm data.

```python
plt.xlabel("Wavelength (nm)")
plt.ylabel("Spectral irradiance W m^-2 nm^-1")
plt.plot(df_zero_air['Wavelength, nm'], df_zero_air['Wehrli, W/m2/nm'])
plt.savefig('ref_zero_air_spectrum.png')
```

Using the specified m data.

```python
plt.xlabel("Wavelength (m)")
plt.xlim(0,4)
plt.ylabel("Spectral irradiance W m^-2 m^-1")
plt.plot(df_zero_air['Wavelength, microns'], df_zero_air['E-490 W/m2/micron' ])
plt.savefig('ref_zero_air_spectrum.png')
```

```python
plt.xlabel("Wavelength (nm)")
plt.xlim(0,4000)
plt.xticks(range(0, 4000, 500))
plt.ylabel("Spectral irradiance W m^-2 nm^-1")
plt.plot(
    [ x * 10**3 for x in df_zero_air['Wavelength, microns'] ],
    [ x * 10**-3 for x in df_zero_air['E-490 W/m2/micron' ] ]
)
#plt.plot(df_zero_air['Wavelength, nm'], df_zero_air['Wehrli, W/m2/nm'])
plt.savefig('ref_zero_air_spectrum.png')
```
